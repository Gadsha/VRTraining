```python
#Task accomplishment time

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(10)

# Generate random data
#data_to_plot = pd.DataFrame(data=np.random.rand(100, 9) * 1000, columns=list('ABCDEFGHI'))
data_to_plot = pd.read_csv('file1.csv')
#data_to_plot.to_csv('file1.csv')
#pd.set_option('display.max_rows', None)
#display(data_to_plot)
# Create a boxplot with Seaborn, setting the box colors to yellow
sns.set(style="white")  # Set the style of the plot to have a white grid background
plt.figure(figsize=(5, 3))  # Set the figure size of the plot
box_plot = sns.boxplot(data=data_to_plot, color='#ff9900',
                       boxprops=dict(edgecolor='none'),
                       medianprops=dict(color='darkred', linestyle='-', linewidth=1),  # Red, thin median line
                       whiskerprops=dict(color='darkred', linestyle='--', linewidth=1), 
                      width=0.5)  # Set the color of the boxes to yellow

# Setting title and labels for axes
box_plot.set_title('Task accomplishment time by different runs, 50 runs')
box_plot.set_xlabel('AI Agent Index')
box_plot.set_ylabel('Task accomplishment time (sec)')
                    
#sns.despine(offset=10, trim=True)
plt.ylim(10, 500)

# Display the plot
plt.savefig('char01.pdf', dpi=100)
plt.show()
```

```python
#Moving distance by AI Agent

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(10)

# Generate random data
#data_to_plot = pd.DataFrame(data=np.random.rand(100, 9) * 1000, columns=list('ABCDEFGHI'))
data_to_plot = pd.read_csv('file3.csv')
#data_to_plot.to_csv('file1.csv')
#pd.set_option('display.max_rows', None)
#display(data_to_plot)
# Create a boxplot with Seaborn, setting the box colors to yellow
sns.set(style="white", palette="pastel")  # Set the style of the plot to have a white grid background
plt.figure(figsize=(5, 3))  # Set the figure size of the plot
box_plot = sns.boxplot(data=data_to_plot, color='#ff9900',
                       boxprops=dict(edgecolor='none'),
                       medianprops=dict(color="darkred"),
                       whiskerprops=dict(color='darkred', linestyle='--'),
                      width=0.5)  # Set the color of the boxes to yellow

# Setting title and labels for axes
box_plot.set_title('Moving distance by different runs, 50 runs')
box_plot.set_xlabel('AI Agent Index')
box_plot.set_ylabel('Moving distance (MeshTiles) / 100')
                    
sns.despine(offset=10, trim=False)
plt.ylim(0, 10)

# Display the plot
plt.savefig('char03.pdf', dpi=100)
plt.show()
```
